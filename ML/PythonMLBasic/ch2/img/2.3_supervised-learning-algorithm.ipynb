### 2.3.1 一些样本数据集

一个模拟的二分类数据集示例是forge数据集

```python
# 生成forge数据集
X, y = mglearn.datasets.make_forge()
# 数据集绘图
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=3)
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.title("forge dataset")
plt.show()
```
